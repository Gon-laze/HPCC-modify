```python
import random
import string
import os
```

```python
#tmpTest
# 生成对应流的flow.txt
flowf = open("flow.txt", "w", encoding="utf-8")
```

```python
topoNode = 5
switchNode = 1
sendNode = 3

dupRound = 3
allow_OverLapping = False
# 如果allow_OverLapping设为True，Timegap_inRound和Timegap_between最好不要都设置为0（否则产生多个相同的重复流）
Timegap_inRound = 0.0
Timegap_between = 0.01

# flowNum = small_flow_num+mid_flow_num+big_flow_num
flowNum = 300
flowf.write(str(flowNum * dupRound)+"\n")

dir_allList = [
    ["high"],
    ["middle"],
    ["low"]
]
```

```python
#dup version
i = 0
for roundIndex in range(0, dupRound):
    for classIndex in range(0, sendNode):
        for dir in dir_allList[classIndex]:
    #         j = 0
            print(dir)
            file_list = [f for f in os.listdir(dir) if os.path.isfile(dir+"/"+f) and f[-4:]==".txt"]
            flsize = len(file_list)
            flprefix = file_list[0].split("_")[0]
            print(flsize)
            print("...")
            print(flprefix)
            for j in range(0, flsize):
                file = open(dir+"/"+flprefix+"_"+str(j)+".txt", "r", encoding="utf-8")
                line = file.readline().split("\n")
                print(line)
                if (allow_OverLapping):
                    # 即使重叠，每轮起始也会轮换————(i + roundIndex)
                    flowf.write(str((i+roundIndex)%sendNode + 1)+" 4 3 100 "+line[0]+" "+str(2 + i*Timegap_inRound + roundIndex*Timegap_between)+"\n")
                else:
                    # 跳过编号为round +1 的节点（该节点用于接收）
                    if (roundIndex > 0):
                        offset = 2
                    else:
                        offset = 1
                    flowf.write(str((i+roundIndex)%sendNode + offset + roundIndex*sendNode)+" 4 3 100 "+line[0]+" "+str(2 + i*Timegap_inRound + roundIndex*Timegap_between)+"\n")
                
                i += 1
                file.close() 
```

```python

flowf.close()
```
