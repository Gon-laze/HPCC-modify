```python
import os
```

```python
dir_list = [dir for dir in os.listdir("./") if os.path.isdir(dir) and dir != ".ipynb_checkpoints"]
dir_list
```

```python
for dir in dir_list:
    count = 0;
    file_list = [f for f in os.listdir(dir) if os.path.isfile(dir+"/"+f) and f[-4:]==".txt"]
    print(file_list)
    for f in file_list:
        os.rename(dir+"/"+f, dir+"/"+dir+"_"+str(count)+".txt")
        count += 1
```
